```python
#● Genere funciones para calcular la media y varianza de un vector. Debe cumplir con los siguientes requisitos:
#○ Ambas funciones deben ingresar un argumento x correspondiente al vector.
#○ Las funciones deben contener docstrings con la documentación asociada a la variable y retornar un valor utilizando return.
#○ La función de la varianza se debe llamar a la función de la media.
# Utilice las funciones para reportar la información sobre goles_favor, goles_contra y puntos.

def mean_vector(x):
    '''
mean_rows(): Funcion encargada de calcular el valor promedio de una columna en particular
x: Variable encargada de pasar la columna a la funcion
x debe resivir solo valores numericos
return tmp_vector: Retorno de variable que almacena el resultado de la funcion
    '''
    x = x.dropna()
    tmp_mean = np.mean(x)# Elimina valores nulos o perdido, limpia los datos
    print(f'El promedio es: {tmp_mean}')
    return tmp_mean

def var_vector(x):
    '''
var_vector(): Funcion encargada de calcular la varianza de una columna en particular
x: Variable encargada de pasar la columna a la funcion
x debe resivir solo valores numericos
return tmp_var, tmp_mean: Retorno de variables que almacenan los resultados de la funcion
    '''
    x = x.dropna()# Elimina valores nulos o perdido, limpia los datos
    tmp_var = np.var(x)
    tmp_mean = np.mean(x)
    print(f'La varianza es: {tmp_var}\nLa media es: {tmp_mean}')
    return tmp_var, tmp_mean
```

```python
mean_gf= mean_vector(df['goles_favor'])
mean_gc = mean_vector(df['goles_contra'])
mean_p = mean_vector(df['puntos'])
print(f'El promedio de goles a favor es: {mean_gf}\nEl promedio de goles en contra es: {mean_gc}\nEl promedio de puntos es: {mean_p}')
```

```python
var_gf= var_vector(df['goles_favor'])
var_gc = var_vector(df['goles_contra'])
var_p = var_vector(df['puntos'])
print(f'La varianza de goles a favor es: {var_gf}\nLa varianza de goles en contra es: {var_gc}\nLa varianza de puntos es: {var_p}')
```

```python
def edad():
    return np.random.choice(['young_pet', 'old_pet'])

simulate_pets_prob=[]

for i in range(20):
    simulate_pets_prob.append(edad())
```

```python
def simulate_pet(x):
    np.random.seed(1)
    prob_uno=0
    prob_dos=0
    prob_tres=0
    for i in range(x):
        a=generate_pet()
        b=edad()
        c=generate_pet()
        d=edad()
        if (a=="perro" or c=="perro"):
            prob_uno+=1
        if((a=="perro" and b=="old_pet")or (c=="perro" and d=="old_pet")):
            prob_dos+=1
        if (a=="perro" or c=="perro"):
            prob_tres+=1
    prob_1 = (prob_uno / x) * 100
    prob_2 = (prob_dos / x) * 100
    prob_3 = (prob_tres / x)* 100
    print(f'''
    El ecenario mas probable es de {prob_1} % 
    El ecenario menos probable es de {prob_2} % 
    Y la probabilidad maxima es de {prob_3} %''')
    
    return prob_1, prob_2, prob_3
```
