```python
def describe_rows(x):
    x = x.dropna()
    tmp_std = np.std(x)
    tmp_mean = np.mean(x)
    tmp_median = np.median(x)
    tmp_var = np.var(x)
    tmp_max = tmp_mean + tmp_std
    tmp_min = tmp_mean - tmp_std
    tmp_mode = stats.mode(x)[0][0]
    print(f'El valor minimo es :{tmp_min}\nEl valor maximo es: {tmp_max}\nEl promedio es: {tmp_mean}\nLa desviacion estandart es: {tmp_std}\nLa moda es: {tmp_mode}\nLa varianza es: {tmp_var}')
    return tmp_mean, tmp_median, tmp_var, tmp_mode, tmp_std, tmp_min, tmp_max
```

```python
describe_rows(df['puntos'])
describe_rows(df['juegos_ganados'])
gd = df['goles_diferencia']
print(gd)
```

```python
#● ¿Cuál es la probabilidad de elegir un equipo asiático al azar? calculo por puntos ganados
media_gen = df['puntos'].mean()
std_puntos = df['puntos'].std()
min = df['puntos'].min()
max = df['puntos'].max()

escoje_asia_puntos = 0
for rowname, rowserie in df.iterrows():
    if rowserie['continent'] == 'asia' and rowserie['puntos'] > min and rowserie['puntos'] < media_gen:
        print(rowserie['team'])
        escoje_asia_puntos += 1
print(escoje_asia_puntos)
```

```python
#● ¿Cuál es la probabilidad de elegir un equipo asiático al azar?
# Hay una probabilidad del 12% de escojer un equipo asiatico al azar segun sus puntos ganados
print(df['escoje_asia_puntos'].value_counts())
print(df['escoje_asia_puntos'].value_counts('%'))
porcentaje_asia = df['escoje_asia_puntos'].value_counts('%')
print(f'la probabilidad de escojer un equipo asiatico al azar es: {porcentaje_asia[1]*100} %')
```
