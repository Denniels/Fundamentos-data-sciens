```python
#definir estadisticos de prueba
trust = df['wvs_trust'].dropna()
diff = trust.mean() - .7
diff
```

```python
errro_std = np.sqrt(np.var(trust, ddof = 1) / float(trust.shape[0]))
errro_std
```

```python
diff / errro_std
```

```python
stats.ttest_1samp(trust, .7)
```
