```python
scattr_data = df.loc[:, ['undp_hdi', 'icrg_qog', 'ccodealp']].dropna()
scattr_data.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(x = scattr_data['undp_hdi'], y = scattr_data['icrg_qog'], marker = '.', s = 1)
for i in scattr_data.index:
    ax.text(scattr_data.loc[i,'undp_hdi'],scattr_data.loc[i,'icrg_qog'],str(scattr_data.loc[i,'ccodealp']))
```

```python
pendiente, intercepto = np.polyfit(scattr_data['undp_hdi'],scattr_data['icrg_qog'],1)
print(f'''
{pendiente}
{intercepto}
''')
```

```python
linear_fit = [pendiente * j + intercepto for j in scattr_data['undp_hdi']]
```

```python
fig, ax = plt.subplots()
ax.scatter(x = scattr_data['undp_hdi'], y = scattr_data['icrg_qog'], marker = '.', s = 1)
ax.plot(scattr_data['undp_hdi'], linear_fit, color = 'tomato')
for i in scattr_data.index:
    ax.text(scattr_data.loc[i,'undp_hdi'],scattr_data.loc[i,'icrg_qog'],str(scattr_data.loc[i,'ccodealp']))
```
