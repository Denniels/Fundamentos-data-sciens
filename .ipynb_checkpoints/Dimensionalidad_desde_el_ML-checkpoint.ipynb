Analicis de componentes principales

el asnalicis de componentes prioncipales es un metodo de transformacion lineal ortogonal que mapea datos en un nuevo sistema de coordenadas, maximizando la varianza 
de las extrapolaciones.

$$ \tilde X = P \times X$$

Donde:
* $\tilde X$: Matriz de datos redimensionada.
* P: Transformacion Lineal: rango y rotacion.
* X: Matriz de datos original

El Análisis de Componentes Principales busca reducir la redundancia de los datos entre una
serie de expansiones basales candidatas. Aquella que aproxime su matriz de covarianza a
una matriz diagonal (entradas diagonales cercanas a 1 y triángulos cercanos a 0), se
denomina como el primer componente principal.

El proceso se itera, rankeando todas las expansiones basales en términos de maximización
de varianza y reducción de redundancia de datos. El proceso se ilustra en la figura
gfx.pca_variance().


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import factor_analyzer as factor
import missingno as msngo
import scipy.stats as stats
import seaborn as sns
import lec7_graphs as gfx
import warnings
warnings.filterwarnings(action = "ignore")
plt.style.use('seaborn')
```

```python
from sklearn.datasets import load_digits
df = load_digits()
```

```python
df.data[8].reshape([8, 8])
```

```python
plt.imshow(df.data[8].reshape([8, 8]), cmap = 'bone')
```

```python
print("Dimensión original de la base:",df.data.shape)
```

```python
from sklearn.decomposition import PCA
dimred = PCA(n_components = 2)
project_x = dimred.fit_transform(df.data)
```

```python
df.data.shape
```

```python
project_x.shape
```

```python
plt.scatter(project_x[:, 0], project_x[:, 1], c = df.target, edgecolor='none', alpha=.7, cmap='Set1')
plt.colorbar()  
```

```python
print(f'La varianza explicada de cada dimensión es de: {dimred.explained_variance_ratio_}')
print(f'En conjunto, ambas dimensiones explican la varianza en un: {sum(dimred.explained_variance_ratio_)}')
```

```python
fetch_dims = PCA().fit(df.data)
dims_quant = len(fetch_dims.components_)
plt.bar(range(1, dims_quant + 1), fetch_dims.explained_variance_ratio_,label = 'Varianza de cada dimensión')
plt.step(range(1, len(fetch_dims.components_) + 1), np.cumsum(fetch_dims.explained_variance_ratio_), color = 'tomato', label = 'Varianza acumulada')
plt.xlabel('Número de dimensiones')
plt.ylabel('Varianza')
plt.legend()
```

```python
pca = PCA(n_components = 10)
project_x = pca.fit_transform(df.data)
print("Varianza acumulada con 10 dimensiones:", np.cumsum(pca.explained_variance_ratio_)[-1])
```
