```python
from operator import index


cantidad_pasajeros = 0
years = []
for columna, serie in df.iteritems():
    if columna == 'year':
        years.append(serie.index.values)

    if columna == 'passengers':
        if serie.index.values in years:
            cantidad_pasajeros += serie.values

print(cantidad_pasajeros)
```
